## 1) Key Points

**Artificial Intelligence (AI)** — Systems that perform tasks that typically require human intelligence (perception, learning, reasoning, decision‑making, language, planning).

**Weak AI**
* also known as Artificial Narrow Intelligence (ANI).
* Designed to perform a specific task.
* Example: Alexa, Siri, Google Assistant.

**History**
*   John McCarthy (1956) → Coined the term Artificial Intelligence.
*   Arthur Samuel (1959) → Coined the term Machine Learning.


**Machine Learning (ML)** — A subfield/subset of AI that learns patterns from data to make predictions/decisions **without** being explicitly programmed with task‑specific rules.
(Examples: regression, classification, clustering, deep learning.)  


**Machine Learning Workflow**
1.   **Data** — Collection of raw information.
2.   **Training the Machine** — Feeding data into algorithms.
3.   **Building a Model** — Model learns from the data.
4.   **Predicting Outcomes** — Using the trained model for predictions.

**Machine Learning Life Cycle:**
1. Data Collection
2. Data Wrangling (Cleaning & Preprocessing)
3. Data Analysis
4. Model Training
5. Model Evaluation
6. Deployment
7. Feedback & Monitoring


**NumPy**:
* NumPy = Numerical Python
* Core library for numeric and scientific computing in Python.
* Supports multi-dimensional arrays and mathematical operations

**Pandas**:
* Pandas = Panel Data
* Library used for data manipulation and analysis.
* Provides two primary data structures:Series (1D) and DataFrame (2D)

**Matplotlib (pyplot)** — Low‑level plotting library with fine control of figures/axes; `pyplot` is its MATLAB‑like interface.  

**Seaborn** — High‑level statistical visualization on top of Matplotlib; works directly with tidy Pandas DataFrames (smart defaults, palettes, faceting).

```python
x1=np.arange(1,11)
y1=2*x1
y2=3*x1
```

```python
plt.plot(x1,y1,color="g",linestyle=":",linewidth="5")
plt.plot(x1,y2,color="r",linestyle="-.",linewidth="2")
plt.title("Line Plot consisting of 2 lines")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
#subplots
plt.subplot(1,2,1) #nrow,ncols,index
plt.plot(x1,y1,color="g",linestyle=":",linewidth="5")
plt.title("Line Plot 1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x1,y2,color="r",linestyle="-.",linewidth="2")
plt.title("Line Plot 2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)

plt.show()
```

```python
#subplots
plt.subplot(2,1,1) #nrow,ncols,index
plt.plot(x1,y1,color="g",linestyle=":",linewidth="5")
plt.title("Line Plot 1")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)

plt.subplot(2,1,2)
plt.plot(x1,y2,color="r",linestyle="-.",linewidth="2")
plt.title("Line Plot 2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)

plt.show()
```

tips dataset

```python
tips = sns.load_dataset("tips")
display(tips.head())
print("shape:", tips.shape)
```

```python
# Scatter (relplot): total_bill vs tip
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter")
plt.title("Scatter: total_bill vs tip")
plt.show()
```

```python
# Scatter with hue = smoker
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", kind="scatter")
plt.title("Scatter with hue=smoker")
plt.show()
```

```python
# Scatter with hue & style = smoker
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", style="smoker", kind="scatter")
plt.title("Scatter: hue & style = smoker")
plt.show()
```

```python
# Scatter with hue=smoker and style=time
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", style="time", kind="scatter")
plt.title("Scatter: hue=smoker, style=time")
plt.show()
```

```python
# Scatter with continuous hue (size)
sns.relplot(data=tips, x="total_bill", y="tip", hue="size", kind="scatter", palette="viridis")
plt.title("Scatter: hue=size (continuous)")
plt.show()
```

```python
sns.relplot(
data=tips, x="total_bill", y="tip",
size="size", sizes=(15, 200)
)
```

```python
sns.relplot(

data=tips,
x="total_bill", y="tip", hue="smoker", col="time",

)
```

```python
# Histogram (histplot)
sns.histplot(data=tips, x="total_bill", bins=20, kde=True)
plt.title("Histogram: total_bill with KDE")
plt.show()
```

```python
# Multiple histograms overlay by smoker
sns.histplot(data=tips, x="tip", hue="smoker", bins=20, element="step", stat="density", common_norm=False)
plt.title("Histogram: tip by smoker (density)")
plt.show()
```

```python
# Bar: mean tip by day
sns.barplot(data=tips, x="day", y="tip", estimator=np.mean, ci="sd")
plt.title("Bar: Mean tip by day (±SD)")
plt.show()
```

```python
# Count plot: number of bills by day/time
sns.countplot(data=tips, x="day", hue="time")
plt.title("Count: bills by day & time")
plt.show()
```

```python
# Boxplot
sns.boxplot(data=tips, x="day", y="tip", hue="smoker")
plt.title("Boxplot: tip by day & smoker")
plt.show()
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="box")
plt.show()
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="box",hue="smoker")
plt.show()
```

```python
sns.catplot(data=tips,x="day",y="total_bill")
plt.show()
```

```python
#jitter= to control magnitude
sns.catplot(data=tips,x="day",y="total_bill",jitter=False)
plt.show()
```

```python
#swarm plot = to prevent overlapping
sns.catplot(data=tips,x="day",y="total_bill",kind="swarm")
plt.show()
```

```python
#swarm plot = to prevent overlapping
sns.catplot(data=tips,x="day",y="total_bill",kind="swarm",hue="sex")
plt.show()
```

```python
#order paarmeter= to display multiple categorical plots in same figure
sns.catplot(data=tips,x="smoker",y="tip",order=["No","Yes"])
plt.show()
```

```python
#categorical plot on vertical axis
sns.catplot(data=tips,x="total_bill",y="day",kind="swarm",hue="time")
plt.show()
```

```python
#Violin Plot
sns.catplot(data=tips,x="total_bill",y="day",kind="violin",hue="sex")
plt.show()
```

```python
sns.catplot(data=tips,x="total_bill",y="day",kind="violin",hue="sex",split=True)
plt.show()
```
